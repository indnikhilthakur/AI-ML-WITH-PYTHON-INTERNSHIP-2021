# Read data from JSON

```python
import json
fd = open("inventory1.0.json", "r")
r = fd.read()
fd.close()

inventory = json.loads(r)
```

```python
inventory
```

# Add multiple items to inventory

```python
def new_record(x):
    for i in range(x):
        item_id = input("item_id")
        name = input("name")
        price = int(input("price"))
        catagory = input("catagory")
        quality = int(input("quality"))    #1 to 5 :1<5
        quantity = int(input("quantity"))
        inventory[item_id] = {'name' : name, 'catagory' : catagory, 'quality' : quality, 'price' : price, 'quantity' : quantity}
    
        js = json.dumps(inventory)
        fd = open('inventory1.0.json', 'w')
        fd.write(js)
        fd.close()
        
    return inventory
    
```

```python
new_record(1)
```
